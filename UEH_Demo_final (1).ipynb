# Tiền xử lý dữ liệu HỌC SINH sử dụng VnCoreNLP

```python
import pandas as pd
import json
from vncorenlp import VnCoreNLP
import re

# --- KHỞI TẠO VnCoreNLP ---
vncorenlp_path = "vncorenlp/VnCoreNLP-1.1.1.jar" # Gán đường dẫn để sử dụng VNcoreNLP
try:
    rdrsegmenter = VnCoreNLP(vncorenlp_path, annotators="wseg", max_heap_size='-Xmx2g') # khởi tạo công cụ tách từ (annotators="wseg") từ VnCoreNLP
    print("Khởi tạo VnCoreNLP thành công!")
except Exception as e:
    print(f"Lỗi khi khởi tạo VnCoreNLP: {e}")
    rdrsegmenter = None

# --- ĐỌC DỮ LIỆU TỪ FILE hoc_sinh_gia_lap.csv RA DATAFRAME ĐỂ THAO TÁC ---
df_hoc_sinh = pd.read_csv('hoc_sinh_gia_lap.csv')


    # --- DANH SÁCH STOPWORDS TIẾNG VIỆT ---
vietnamese_stopwords = [
        "và", "là", "của", "trong", "được", "cho", "có", "không", "để", "khi", "thì", "mà", "với",
        "một", "các", "những", "này", "đó", "khác", "từ", "đến", "về", "như", "ở", "tại",
        "cũng", "sẽ", "đã", "đang", "hơn", "rất", "qua", "vào", "ra", "lên", "xuống", "sau",
        "học", "sinh", "viên", "chương trình", "đào tạo", "kiến thức", "kỹ năng", "ngành",
        "công việc", "vị trí", "doanh nghiệp", "tổ chức", "công ty", "khả năng", "mục tiêu",
        "yêu cầu", "giải quyết", "vấn đề", "thực tiễn", "lĩnh vực", "phát triển", "xây dựng",
        "hoạt động", "quản lý", "ứng dụng", "hệ thống", "thông tin", "quốc tế", "toàn cầu",
        "số", "thứ", "điều", "việc", "cách", "bản", "thân", "đối", "tượng", "mọi", "cùng",
        "1", "2", "3", "i", "ii", "iii",
        "ueh", "trường", "đại học", "kinh tế", "tp hcm", "hồ chí minh", "tốt nghiệp",
        "sau khi hoàn thành", "chính quy", "chuẩn", "cử nhân",
        "mức độ tự chủ và trách nhiệm", "tự chủ", "trách nhiệm", "thái độ",
        "cơ hội việc làm", "cơ hội nghề nghiệp", "vị trí việc làm", "mô tả công việc",
        "organization", "position", "job description",
        "hiểu", "nhận biết", "giải thích", "vận dụng", "áp dụng", "phân tích", "tóm lược", "đánh giá",
        "có", "đạt", "được", "thể hiện", "nắm vững", "thuần thục", "tham gia", "thực hiện",
        "theo", "bao gồm", "cũng như", "thuộc", "liên quan", "nhằm", "để", "phục vụ",
        "ngoại ngữ", "tin học", "giáo dục thể chất", "giáo dục quốc phòng", "khung năng lực", "bậc",
        "quyết định", "ngày", "tháng", "năm", "qđ-đhkt-đbclptct", "qđ-đhkt-nn"
        "1", "2", "3", "4", "5", "6", "7", "8", "9", "0",
        "i", "ii", "iii", "iv", "v", "vi", "vii", "viii", "ix", "x"
    ]

stopwords_set = set(vietnamese_stopwords)

# --- HÀM TIỀN XỬ LÝ SỬ DỤNG VnCoreNLP ---
def preprocess_text_vncorenlp(text, segmenter, stopwords):
    if not isinstance(text, str) or not text.strip() or not segmenter: # Kiểm tra đầu vào: nếu không phải chuỗi hoặc rỗng hoặc không có mô hình, trả về chuỗi rỗng.
        return ""

    # Làm sạch văn bản
    text = re.sub(r'\s+', ' ', text) # thay thế nhiều khoảng trắng bằng 1 khoảng trắng
    text = re.sub(r'[^\w\s]', '', text) # loại bỏ ký tự đặc biệt
    text = text.lower() # chuyển về chữ thường

    # Tách từ sử dụng mô hình VnCoreNLP, sau đó "làm phẳng" danh sách từ ra 1 chiều để xử lý từng từ riêng biệt
    try:
        segmented_sentences = segmenter.tokenize(text)
        words = [token for sentence_tokens in segmented_sentences for token in sentence_tokens]
    except Exception as e_seg:
        print(f"Lỗi khi tách từ cho text: '{text[:50]}...' - Lỗi: {e_seg}")
        words = text.split()

    keywords = [word for word in words if word not in stopwords and len(word) > 1 and not word.isdigit()]
    return " ".join(keywords)

if rdrsegmenter: # Chỉ chạy nếu VnCoreNLP được khởi tạo thành công
    print("\nĐang tiền xử lý các cột văn bản bằng VnCoreNLP...")

    cols_to_process = ['so_thich', 'ky_nang']  # Chỉ định các cột văn bản cần tiền xử lý
    # Tiền xử lý từng cột văn bản đã chỉ  định
    for col_to_process in cols_to_process:  # duyệt qua các cột đã chỉ định
        if col_to_process in df_hoc_sinh.columns:  # nếu tồn tại trong bảng dữ liệu
            processed_col_name = f"{col_to_process}_vncore_processed" # Tạo tên mới cho cột kết quả
            print(f"  Đang xử lý cột: {col_to_process} -> {processed_col_name}")
            # Áp dụng hàm preprocess_text_vncorenlp cho từng dòng:
            df_hoc_sinh[processed_col_name] = df_hoc_sinh[col_to_process].astype(str).apply(
                lambda x: preprocess_text_vncorenlp(x, rdrsegmenter, stopwords_set)
            )
        else:
            print(f"  Cột {col_to_process} không tồn tại để xử lý.")


    print("\n5 dòng đầu của DataFrame sau khi tiền xử lý bằng VnCoreNLP:")
    # Hiển thị các cột đã xử lý cùng tên ngành để dễ kiểm tra
    display_cols = ['ma_hoc_sinh'] + [f"{col}_vncore_processed" for col in cols_to_process]
    print(df_hoc_sinh[display_cols].head())

    # Lưu kết quả
    output_csv_path_vncore = "hoc_sinh_gia_lap_vncore_processed.csv"
    df_hoc_sinh.to_csv(output_csv_path_vncore, index=False, encoding='utf-8-sig')
    print(f"\nDữ liệu đã tiền xử lý bằng VnCoreNLP được lưu vào file: {output_csv_path_vncore}")

else:
    print("\nVnCoreNLP chưa được khởi tạo. Bỏ qua bước tiền xử lý bằng VnCoreNLP.")
```

# CRAWL DỮ LIỆU TÊN NGÀNH + LINK NGÀNH

```python
# Khởi chạy Selenium với Chrome ở chế độ headless
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
import pandas as pd
import re # Import re để sử dụng biểu thức chính quy

## Khởi chạy trình duyệt Chrome ở chế độ không giao diện (headless) để chạy ngầm nhanh hơn:
# Cấu hình WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
# Tạo trình điều khiển
service = Service("/usr/bin/chromedriver") # Đảm bảo đường dẫn này chính xác
driver = webdriver.Chrome(service=service, options=options)


# Mở trang tuyển sinh UEH
url = "https://tuyensinh.ueh.edu.vn/cu-nhan-chinh-quy/dai-hoc-chinh-quy/"
driver.get(url)

wait = WebDriverWait(driver, 20) # chờ đợi tải các phần tử HTML

nganh_list_final = [] # khởi tạo mảng chứa danh sách ngành
current_page = 1
processed_links = set()

# --- TẠO BỘ LỌC ĐƠN GIẢN DỰA VÀO ĐUÔI URL ---
VALID_MAJOR_URL_ENDINGS_REGEX = [
    r"chuong-trinh-chuan/?$",
    r"chuong-trinh-tieng-viet/?$"
] # chỉ lọc ra những ngành kết thúc bằng "chuong-trinh-chuan" hoặc "chuong-trinh-tieng-viet".
# Tạo một biểu thức regex tổng hợp
combined_valid_endings_regex = '|'.join(f"({pattern})" for pattern in VALID_MAJOR_URL_ENDINGS_REGEX)


while True: # duyệt tuần tự từng trang ds ngành
    print(f"\n📄 Đang xử lý trang {current_page}...")

    try:
        main_content_selector = "#dhcq-chuan" # xác định phần tử HTML có ID là dhcq-chuan
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, main_content_selector))) # Đợi DOM tải xong, sau đó lấy danh sách ngành hiển thị trên trang.
        sleep(1) # Dừng chương trình thêm 1 giây sau khi phần tử xuất hiện để hiển thị hoàn toàn nội dung bên trong, tránh lỗi lấy thiếu các ngành.

        items_elements = driver.find_elements(By.CSS_SELECTOR, f"{main_content_selector} .jet-listing-grid__item") # lấy tất cả các item ngành học được chứa trong phần tử có class jet-listing-grid__item
        print(f"🔍 Tìm thấy {len(items_elements)} mục trên trang {current_page}")

        if not items_elements and current_page > 1:
            print("⚠️ Không tìm thấy mục nào trên trang này, có thể đã hết trang hợp lệ.")
            break

        for item_element in items_elements:
            try:
                a_tag = item_element.find_element(By.CSS_SELECTOR, "a")
                link_nganh_raw = a_tag.get_attribute("href").strip()
                link_nganh_lower = link_nganh_raw.lower()

                # --- ÁP DỤNG BỘ LỌC URL ĐƠN GIẢN ---
                if re.search(combined_valid_endings_regex, link_nganh_lower):
                    # Nếu link hợp lệ, lấy tên ngành
                    try:
                        span_tag = item_element.find_element(By.CSS_SELECTOR, "span.elementor-icon-list-text") # Tìm tên ngành trong phần tử span có class elementor-icon-list-text
                        ten_nganh = span_tag.text.strip()
                    except:
                        try:
                            heading_tag = item_element.find_element(By.CSS_SELECTOR, "h2.elementor-heading-title a") # TH khác tên ngành nằm bên trong thẻ <h2 class= "elementor-heading-title"> <a = href...> Tên ngành </a> </h2>
                            ten_nganh = heading_tag.text.strip()
                        except:
                            ten_nganh = "Không lấy được tên ngành (Link: " + link_nganh_raw + ")" # Ghi nhận link nếu không có tên

                    # Kiểm tra trùng lặp link trước khi thêm
                    if link_nganh_lower not in processed_links:
                        nganh_list_final.append({
                            "ten_nganh": ten_nganh,
                            "link_nganh": link_nganh_raw
                        })
                        processed_links.add(link_nganh_lower)
                else:
                    print(f"🗑️ [Lọc URL] Bỏ qua link không khớp đuôi hợp lệ: {link_nganh_raw}")

            except Exception as e_item:
                print(f"❌ Lỗi khi xử lý một mục: {e_item}")

        # Tìm và click nút Next
        try:
            next_btn = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".jet-filters-pagination__item.next")))

            # Kiểm tra nếu không thể nhấn nữa (đã đến trang cuối)
            if "disabled" in next_btn.get_attribute("class"):
                print("✅ Đã đến trang cuối (nút Next bị vô hiệu hóa).")
                break

            # Cuộn để đảm bảo nút hiện ra và click
            driver.execute_script("arguments[0].scrollIntoView(true);", next_btn)
            sleep(0.5)
            driver.execute_script("arguments[0].click();", next_btn)
            sleep(1.5)  # Chờ trang mới
            current_page += 1

        except Exception as e_next:
            print(f"🏁 Không tìm thấy hoặc không click được nút Next (hoặc đã ở trang cuối cùng). Lỗi: {e_next}")
            break

    except Exception as e_page:
        print(f"❌ Lỗi khi xử lý trang {current_page}: {e_page}")
        break

# Đóng trình duyệt
driver.quit()

# Kết quả cuối cùng
print(f"\n🎉 Thu thập được tổng cộng {len(nganh_list_final)} ngành!")
for nganh in nganh_list_final:
    print(f"- {nganh['ten_nganh']}: {nganh['link_nganh']}")
```

# Crawl dữ liệu chi tiết của từng ngành

```python
# Khởi chạy Selenium với Chrome ở chế độ headless
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd

# Cấu hình WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Tạo trình điều khiển
service = Service("/usr/bin/chromedriver")
driver = webdriver.Chrome(service=service, options=options)

# Lặp qua từng ngành đã thu thập
print("\n Đang cào dữ liệu chi tiết của từng ngành...")
for i, nganh in enumerate(nganh_list_final):
    print(f"🚀 Đang xử lý ngành {i+1}/{len(nganh_list_final)}: {nganh['ten_nganh']}")
    try:
      driver.get(nganh['link_nganh'])
      sleep(0.5)  # Cho trang tải đầy đủ

      soup = BeautifulSoup(driver.page_source, "html.parser")

      # === LẤY TÊN NGÀNH (nếu ten_nganh null) ===
      if not nganh.get("ten_nganh") or nganh["ten_nganh"].strip() == "":
          try:
              ten_nganh = soup.find("h3", style="text-align: center;")
              if ten_nganh:
                  nganh["ten_nganh"] = ten_nganh.get_text(strip=True)
          except Exception as e:
              print(f"⚠️ Lỗi khi lấy tên ngành từ thẻ <h3>: {e}")


      # === LẤY CHUẨN ĐẦU RA ===
      chuan_dau_ra = []
      try:
          chuan_div = soup.find("div", id="jet-toggle-content-4595")
          if chuan_div:
              ul_tags = chuan_div.find_all("ul")
              for ul in ul_tags:
                  for li in ul.find_all("li"):
                      chuan_dau_ra.append(li.get_text(strip=True))
      except Exception as e:
          print(f"⚠️ Lỗi lấy chuẩn đầu ra ngành {nganh['ten_nganh']}: {e}")

      # === LẤY CƠ HỘI VIỆC LÀM ===
      co_hoi_viec_lam = []
      try:
        viec_lam_div = soup.find("div", id="jet-toggle-content-4596")
        if viec_lam_div:
            table = viec_lam_div.find("table")
            if table:
                # NẾU CÓ BẢNG: Trích xuất và gộp text từ các ô của mỗi hàng
                  rows = table.find_all("tr")
                  for row in rows[1:]:  # Bỏ tiêu đề
                      cols = row.find_all("td")
                      if len(cols) >= 3:
                          vi_tri = cols[1].get_text(strip=True)
                          mo_ta = cols[2].get_text(strip=True)
                          co_hoi_viec_lam.append({"vi_tri": vi_tri, "mo_ta": mo_ta})
            else:
                # NẾU KHÔNG CÓ BẢNG: Lấy toàn bộ text từ khối div đã tìm được
                co_hoi_viec_lam = viec_lam_div.get_text(separator='\n', strip=True)

      except Exception as e:
          print(f"⚠️ Lỗi lấy việc làm ngành {nganh['ten_nganh']}: {e}")

      # Gán vào dict ngành
      nganh["chuan_dau_ra"] = chuan_dau_ra
      nganh["co_hoi_viec_lam"] = co_hoi_viec_lam

    except Exception as e:
        print(f"❌ Lỗi khi xử lý ngành {nganh['ten_nganh']}: {e}")
# Xem kết quả mẫu
for nganh in nganh_list_final[:3]:
    print(f"\n📌 Ngành: {nganh['ten_nganh']}")
    print("🔹 Chuẩn đầu ra:")
    for cd in nganh["chuan_dau_ra"]:
        print(f" - {cd}")

    print("🔸 Cơ hội việc làm:")
    for cv in nganh["co_hoi_viec_lam"]:
        print(f" - {cv['vi_tri']}: {cv['mo_ta']}")

# Lưu ra file để dùng tiếp
df = pd.DataFrame(nganh_list_final)
df.to_json("nganh_ueh_chi_tiet.json", force_ascii=False, indent=2)
print("\n💾 Đã lưu thông tin chi tiết vào: nganh_ueh_chi_tiet.json")
```

```python
import pandas as pd
import json
from vncorenlp import VnCoreNLP
import re

# --- KHỞI TẠO VnCoreNLP ---
vncorenlp_path = "vncorenlp/VnCoreNLP-1.1.1.jar" # Đảm bảo đường dẫn đúng
try:
    rdrsegmenter = VnCoreNLP(vncorenlp_path, annotators="wseg", max_heap_size='-Xmx2g')
    print("Khởi tạo VnCoreNLP thành công!")
except Exception as e:
    print(f"Lỗi khi khởi tạo VnCoreNLP: {e}")
    rdrsegmenter = None

# --- ĐỌC DỮ LIỆU TỪ FILE JSON VÀ TẠO CÁC CỘT TEXT GỐC ---
file_path_json = 'nganh_ueh_chi_tiet.json'
try:
    with open(file_path_json, 'r', encoding='utf-8') as f:
        data_from_json = json.load(f)

    num_rows = len(data_from_json['ten_nganh']) # Giả sử 'ten_nganh' luôn tồn tại và quyết định số dòng
    processed_data_list = []
    for i in range(num_rows):
        row_dict = {}
        idx_str = str(i)

        row_dict['TenNganh'] = data_from_json['ten_nganh'].get(idx_str)
        row_dict['URL'] = data_from_json['link_nganh'].get(idx_str) # Giả sử có cột link_nganh

        # Xử lý cột 'chuan_dau_ra' (list of strings)
        cdr_list = data_from_json['chuan_dau_ra'].get(idx_str, [])
        chuan_dau_ra_text = "\n".join(cdr_list) if isinstance(cdr_list, list) else ""
        row_dict['ChuanDauRa_FullText'] = chuan_dau_ra_text

        # Xử lý cột 'co_hoi_viec_lam' (list of dictionaries)
        chvl_list = data_from_json['co_hoi_viec_lam'].get(idx_str, [])
        co_hoi_viec_lam_text_parts = []
        if isinstance(chvl_list, list):
            for item_chvl in chvl_list:
                if isinstance(item_chvl, dict):
                    vi_tri = item_chvl.get("vi_tri", "")
                    mo_ta = item_chvl.get("mo_ta", "")
                    # Gộp cả vị trí và mô tả vào một chuỗi lớn hơn cho mỗi cơ hội
                    if vi_tri: co_hoi_viec_lam_text_parts.append(f"{vi_tri}")
                    if mo_ta: co_hoi_viec_lam_text_parts.append(f"{mo_ta}")
        co_hoi_viec_lam_text = "\n".join(co_hoi_viec_lam_text_parts)
        row_dict['CoHoiViecLam_FullText'] = co_hoi_viec_lam_text

        processed_data_list.append(row_dict)

    df_ueh_nganh = pd.DataFrame(processed_data_list)
    print(f"Đã đọc và xử lý thành công file JSON: {file_path_json}")

except FileNotFoundError:
    print(f"Lỗi: Không tìm thấy file {file_path_json}. Vui lòng kiểm tra lại tên file và đường dẫn.")
    df_ueh_nganh = pd.DataFrame()
except Exception as e:
    print(f"Lỗi khi đọc hoặc xử lý file JSON: {e}")
    df_ueh_nganh = pd.DataFrame()


if not df_ueh_nganh.empty:
    print(f"Số lượng ngành ban đầu: {len(df_ueh_nganh)}")
    print("5 dòng đầu của DataFrame (trước khi tiền xử lý):")
    print(df_ueh_nganh[['TenNganh', 'ChuanDauRa_FullText', 'CoHoiViecLam_FullText']].head())

    # --- DANH SÁCH STOPWORDS TIẾNG VIỆT ---
    vietnamese_stopwords = [
        "và", "là", "của", "trong", "được", "cho", "có", "không", "để", "khi", "thì", "mà", "với",
        "một", "các", "những", "này", "đó", "khác", "từ", "đến", "về", "như", "ở", "tại",
        "cũng", "sẽ", "đã", "đang", "hơn", "rất", "qua", "vào", "ra", "lên", "xuống", "sau",
        "học", "sinh", "viên", "chương trình", "đào tạo", "kiến thức", "kỹ năng", "ngành",
        "công việc", "vị trí", "doanh nghiệp", "tổ chức", "công ty", "khả năng", "mục tiêu",
        "yêu cầu", "giải quyết", "vấn đề", "thực tiễn", "lĩnh vực", "phát triển", "xây dựng",
        "hoạt động", "quản lý", "ứng dụng", "hệ thống", "thông tin", "quốc tế", "toàn cầu",
        "số", "thứ", "điều", "việc", "cách", "bản", "thân", "đối", "tượng", "mọi", "cùng",
        "1", "2", "3", "i", "ii", "iii",
        "ueh", "trường", "đại học", "kinh tế", "tp hcm", "hồ chí minh", "tốt nghiệp",
        "sau khi hoàn thành", "chính quy", "chuẩn", "cử nhân",
        "mức độ tự chủ và trách nhiệm", "tự chủ", "trách nhiệm", "thái độ",
        "cơ hội việc làm", "cơ hội nghề nghiệp", "vị trí việc làm", "mô tả công việc",
        "organization", "position", "job description",
        "hiểu", "nhận biết", "giải thích", "vận dụng", "áp dụng", "phân tích", "tóm lược", "đánh giá",
        "có", "đạt", "được", "thể hiện", "nắm vững", "thuần thục", "tham gia", "thực hiện",
        "theo", "bao gồm", "cũng như", "thuộc", "liên quan", "nhằm", "để", "phục vụ",
        "ngoại ngữ", "tin học", "giáo dục thể chất", "giáo dục quốc phòng", "khung năng lực", "bậc",
        "quyết định", "ngày", "tháng", "năm", "qđ-đhkt-đbclptct", "qđ-đhkt-nn"
        "1", "2", "3", "4", "5", "6", "7", "8", "9", "0",
        "i", "ii", "iii", "iv", "v", "vi", "vii", "viii", "ix", "x"
    ]
    stopwords_set = set(vietnamese_stopwords)

    # --- HÀM TIỀN XỬ LÝ SỬ DỤNG VnCoreNLP ---
    def preprocess_text_vncorenlp(text, segmenter, stopwords):
        if not isinstance(text, str) or not text.strip() or not segmenter:
            return ""
        text = re.sub(r'\s+', ' ', text)
        text = re.sub(r'[^\w\s]', '', text)
        text = text.lower()
        try:
            segmented_sentences = segmenter.tokenize(text)
            words = [token for sentence_tokens in segmented_sentences for token in sentence_tokens]
        except Exception as e_seg:
            print(f"Lỗi khi tách từ cho text: '{text[:50]}...' - Lỗi: {e_seg}")
            words = text.split()

        keywords = [word for word in words if word not in stopwords and len(word) > 1 and not word.isdigit()]
        return " ".join(keywords)

    if rdrsegmenter: # Chỉ chạy nếu VnCoreNLP được khởi tạo thành công
        print("\nĐang tiền xử lý các cột văn bản bằng VnCoreNLP...")

        cols_to_process = ['ChuanDauRa_FullText', 'CoHoiViecLam_FullText']
        # Tiền xử lý từng cột văn bản bạn muốn
        for col_to_process in cols_to_process:
            if col_to_process in df_ueh_nganh.columns:
                processed_col_name = f"{col_to_process}_vncore_processed"
                print(f"  Đang xử lý cột: {col_to_process} -> {processed_col_name}")
                df_ueh_nganh[processed_col_name] = df_ueh_nganh[col_to_process].astype(str).apply(
                    lambda x: preprocess_text_vncorenlp(x, rdrsegmenter, stopwords_set)
                )
            else:
                print(f"  Cột {col_to_process} không tồn tại để xử lý.")


        print("\n5 dòng đầu của DataFrame sau khi tiền xử lý bằng VnCoreNLP:")
        # Hiển thị các cột đã xử lý cùng tên ngành để dễ kiểm tra
        display_cols = ['TenNganh'] + [f"{col}_vncore_processed" for col in cols_to_process]
        print(df_ueh_nganh[display_cols].head())

        # Lưu kết quả
        output_csv_path_vncore = "ueh_nganh_hoc_data_individual_cols_vncore_processed.csv"
        df_ueh_nganh.to_csv(output_csv_path_vncore, index=False, encoding='utf-8-sig')
        print(f"\nDữ liệu đã tiền xử lý bằng VnCoreNLP được lưu vào file: {output_csv_path_vncore}")

    else:
        print("\nVnCoreNLP chưa được khởi tạo. Bỏ qua bước tiền xử lý bằng VnCoreNLP.")

else:
    print("Lỗi: DataFrame rỗng. Không thể tiếp tục.")
```

# MATCHING DỮ LIỆU NGÀNH VÀ DỮ LIỆU HỌC SINH

```python
#  CÀI THƯ VIỆN

!pip install pandas scikit-learn

import pandas as pd
import random
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files

#  ĐỌC DỮ LIỆU NGÀNH HỌC ĐÃ XỬ LÝ NLP
file_nganh = "/content/ueh_nganh_hoc_data_individual_cols_vncore_processed.csv"
nganh_df = pd.read_csv(file_nganh)

# Gộp 2 cột NLP thành full_text
nganh_df['full_text'] = nganh_df['ChuanDauRa_FullText_vncore_processed'].fillna('') + ' ' + \
                        nganh_df['CoHoiViecLam_FullText_vncore_processed'].fillna('')

# ĐỌC DỮ LIỆU HỌC SINH ĐÃ XỬ LÝ NLP
df_hs_processed = pd.read_csv('hoc_sinh_gia_lap_vncore_processed.csv')

#  VECTOR HÓA
vectorizer = CountVectorizer()
nganh_vectors = vectorizer.fit_transform(nganh_df['full_text'])
sothich_vectors = vectorizer.transform(df_hs_processed['so_thich_vncore_processed'])
kynang_vectors = vectorizer.transform(df_hs_processed['ky_nang_vncore_processed'])

#  TÍNH COSINE SIMILARITY
cosine_sothich = cosine_similarity(sothich_vectors, nganh_vectors)
cosine_kynang = cosine_similarity(kynang_vectors, nganh_vectors)

# : TÍNH MATCHING SCORE
matching_score = 0.6 * cosine_sothich + 0.4 * cosine_kynang

# TẠO BẢNG FULL MATCHING
full_results = []
for i, row in enumerate(matching_score):
    for idx in range(len(nganh_df)):
        full_results.append({
            'ma_hoc_sinh': df_hs_processed.iloc[i]['ma_hoc_sinh'],
            'so_thich': df_hs_processed.iloc[i]['so_thich'],
            'ky_nang': df_hs_processed.iloc[i]['ky_nang'],
            'nganh': nganh_df.iloc[idx]['TenNganh'],
            'matching_score': round(row[idx], 4)
        })

matching_full_df = pd.DataFrame(full_results)

#  TRÍCH TOP 5 NGÀNH PHÙ HỢP
top5_df = (
    matching_full_df
    .sort_values(['ma_hoc_sinh', 'matching_score'], ascending=[True, False])
    .groupby('ma_hoc_sinh')
    .head(5)
    .reset_index(drop=True)
)
top5_df['xep_hang'] = top5_df.groupby('ma_hoc_sinh').cumcount() + 1
print("🎯 Gợi ý ngành học phù hợp nhất cho từng sinh viên:")
top5_df.head(20)
```

```python
# Lưu cả 2 bảng
matching_full_df.to_csv("/content/matching_full.csv", index=False, encoding='utf-8-sig')
top5_df.to_csv("/content/matching_top5.csv", index=False, encoding='utf-8-sig')
```

# VISUALIZATION

Cell 1: Bảng tư vấn ngành phù hợp cho 1 học sinh cụ thể và biểu đồ thanh (cho 1 học sinh)


```python
import pandas as pd
import plotly.express as px
import os # Dùng cho việc tạo thư mục

# Cấu hình đường dẫn lưu file
# Đảm bảo thư mục này được tạo một lần
output_dir = "visualization_outputs"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Chọn một học sinh cụ thể để hiển thị
target_student_id = 'SV0000' # Bạn có thể thay đổi ID này

# Lọc dữ liệu cho học sinh cụ thể
student_top_majors = top5_df[top5_df['ma_hoc_sinh'] == target_student_id].copy()

print(f"\n--- Bảng tư vấn ngành phù hợp cho học sinh {target_student_id} ---")

# Hiển thị bảng tư vấn
print(student_top_majors[['xep_hang', 'nganh', 'matching_score']].to_string(index=False))

# Biểu đồ thanh thể hiện matching score với từng ngành (Plotly) cho 1 học sinh
fig_bar = px.bar(
    student_top_majors,
    x='nganh',
    y='matching_score',
    color='xep_hang', # Màu sắc theo xếp hạng
    title=f'Matching Score của Học sinh {target_student_id} với các Ngành (Top 5)',
    labels={'nganh': 'Tên Ngành', 'matching_score': 'Matching Score'},
    hover_data={'xep_hang': True, 'matching_score': ':.4f'},
    text='matching_score' # Hiển thị giá trị trên thanh
)
fig_bar.update_traces(texttemplate='%{text:.4f}', textposition='outside')
fig_bar.update_layout(xaxis_title_text='Ngành Học', yaxis_title_text='Matching Score')
fig_bar.show()
```

Cell 2: Biểu đồ thanh thể hiện matching score với từng ngành (cho TOÀN BỘ học sinh - có lấy mẫu)


```python
import pandas as pd
import plotly.express as px
import numpy as np # Cần numpy cho np.ceil

# Để tránh lag, nhóm sẽ chỉ lấy một mẫu các học sinh để hiển thị.
num_students_to_sample = 20 # Số lượng học sinh muốn hiển thị trong biểu đồ này

sample_students_ids = top5_df['ma_hoc_sinh'].unique()
if len(sample_students_ids) > num_students_to_sample:
    # Lấy top N học sinh đầu tiên để kết quả luôn nhất quán, hoặc bạn có thể dùng random.sample
    sample_students_ids = sample_students_ids[:num_students_to_sample]

top5_df_sample = top5_df[top5_df['ma_hoc_sinh'].isin(sample_students_ids)].copy()

fig_all_students_bar = px.bar(
    top5_df_sample, # Dùng dữ liệu đã được lấy mẫu
    x='nganh',
    y='matching_score',
    color='xep_hang', # Màu sắc theo xếp hạng (1-5)
    facet_col='ma_hoc_sinh', # Chia biểu đồ theo từng học sinh
    facet_col_wrap=min(3, num_students_to_sample), # 3 biểu đồ con trên một hàng, hoặc ít hơn nếu số mẫu nhỏ
    facet_row_spacing=0.005, # Khoảng cách giữa các hàng biểu đồ con (điều chỉnh để tránh lỗi lag)
    title=f'Top 5 Matching Score của Từng Học Sinh (Mẫu {num_students_to_sample} HS) với các Ngành',
    labels={'nganh': 'Tên Ngành', 'matching_score': 'Matching Score', 'xep_hang': 'Xếp hạng'},
    hover_data={'so_thich': True, 'ky_nang': True, 'matching_score': ':.4f'},
    text='matching_score' # Hiển thị giá trị trên thanh
)

fig_all_students_bar.update_traces(texttemplate='%{text:.2f}', textposition='outside')
rows_needed_sample = np.ceil(len(sample_students_ids) / min(3, num_students_to_sample))
# Điều chỉnh chiều cao tổng thể của biểu đồ để phù hợp với số lượng học sinh trong mẫu
fig_all_students_bar.update_layout(height=max(600, int(rows_needed_sample * 250)), showlegend=True)
fig_all_students_bar.for_each_annotation(lambda a: a.update(text=a.text.replace("ma_hoc_sinh=", "Học sinh ")))
fig_all_students_bar.show()
```

Cell 3: Heatmap mức phù hợp giữa các ngành và học sinh


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np # Cần numpy cho unique
import os


# Chọn một số học sinh để hiển thị trên heatmap để biểu đồ không quá dày đặc
# Mặc định chọn 10 học sinh đầu tiên
students_for_heatmap = matching_full_df['ma_hoc_sinh'].unique()[:10]
heatmap_data = matching_full_df[matching_full_df['ma_hoc_sinh'].isin(students_for_heatmap)]

# Tạo pivot table cho heatmap
heatmap_pivot = heatmap_data.pivot_table(index='ma_hoc_sinh', columns='nganh', values='matching_score')

# Sắp xếp các cột ngành theo thứ tự alphabet để hiển thị nhất quán
heatmap_pivot = heatmap_pivot.reindex(sorted(heatmap_pivot.columns), axis=1)

# Heatmap (Seaborn/Matplotlib)
plt.figure(figsize=(18, 9)) # Điều chỉnh kích thước hình cho phù hợp
sns.heatmap(
    heatmap_pivot,
    annot=True,        # Hiển thị giá trị số trên heatmap
    fmt=".2f",         # Định dạng số thập phân (2 chữ số)
    cmap="YlGnBu",     # Bảng màu (có thể thay đổi)
    linewidths=.5,     # Đường kẻ giữa các ô
    cbar_kws={'label': 'Matching Score'} # Nhãn cho thanh màu
)
plt.title('Heatmap Mức Độ Phù Hợp (Matching Score) giữa Học sinh và Ngành')
plt.xlabel('Ngành Học')
plt.ylabel('Mã Học Sinh')
plt.xticks(rotation=45, ha='right') # Xoay nhãn trục x để dễ đọc nếu tên ngành dài
plt.yticks(rotation=0) # Đảm bảo nhãn trục y không bị xoay
plt.tight_layout() # Điều chỉnh layout để tránh cắt xén nhãn
plt.show()
```
